Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

```python
def print_documents(documents):
    for doc in documents:
        print(doc)
```

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

```python
def print_directories(directories):
    for key,value in directories.items():
        print(key,value)


def searchNumDoc(numbdocum):
    for fow in documents:
        if str(numbdocum) in fow['number']:
            return fow['name']
    return None

def searchNumDocInDirect(numbdocum):
    for direct in directories.items():
        if str(numbdocum) in direct[1]:
            return direct[0]
    return None


def searchPeopleByDoc(numbdocum): #p – people 
    resultSearch = searchNumDoc(numbdocum)
    if resultSearch == None:
        return 'такого документа/человека нет'
    else:
        return resultSearch
    




def searchShelfByDoc(numbdocum): #s – shelf
    resultSearch = searchNumDocInDirect(numbdocum)
    if resultSearch == None:
        return 'такого документа/полки нет'
    else:
        return resultSearch
    




def outListAll(): #l-list
    outSring = ''
    valuesArr = [(value['type'], value['number'], value['name']) for value in documents]
    for value in valuesArr:
        outSring += f'{value[0]} "{value[1]}" "{value[2]}"\n'
    return outSring




def checkKey(newKey):
    for key in directories:
        if key == str(newKey):
            return False
    return True

def addShelf(newKey): 
    if checkKey(str(newKey)):
        directories[newKey] = ''
        return 'успешно'
    else:
        return 'уже существует'
    
    
def delDoc(numbdocum): 
    for numDic, doc in enumerate(documents):
        if str(numbdocum) == doc['number']:
            del documents[numDic]
            break
    else:
        return 'документа нет'
    
    for valuesNumDoc in directories.values():
        if str(numbdocum) in valuesNumDoc:
            valuesNumDoc.remove(str(numbdocum))
            
    return 'дкумнт удален'






def checkShelf(numshelf):
    if numshelf not in directories:
        return True
    return False


def moveDocInShelf(numbdocum,numshelf): 
    if checkShelf(numshelf):
        return 'полки нет'
    for key, value in directories.items():
        if numbdocum in value:
            directories[numshelf] += [numbdocum]
            value.remove(numbdocum)
            return 'вы умпешно перенесли документ'
    return 'дкумент не найден'



def addDoc(numbdocum, typedocum, ownerName, numshelf):
    if checkShelf(numshelf):
        return 'полки нет'
    
    newDoc = {'type': typedocum, 'number': numbdocum, 'name': ownerName}
    documents.append(newdocum)
    directories[numshelf] += [numbdocum]
    return 'добавлен документ'




def situation(mode='all'):
    if mode == 'documents': 
        print('документы на данный моомент:')
        print_documents(documents)
    elif mode == 'directories':
        print('щкаф на данный момент:')
        print_directories(directories)
    else:
        print('документы на данный моомент:')
        print_documents(documents)
        print('щкаф на данный момент:')
        print_directories(directories)
        
        
        
        
        
        
        
        
def menu():
    while True:
        userOut = input('введите addadd для заполнения (доступные команды - commands)\n')
        if userOut == 'commands':
            print('//'.join(tipMenu[:-1]))
        elif userOut.lower() in ('p','people'):
            userDoc = input('номер документа:\n')
            print(searchPeopleByDoc(userDoc))
            
        elif userOut.lower() in ('s','shelf'):
            userDoc = input('номер документа:\n')
            print(searchShelfByDoc(userDoc))
            
        elif userOut.lower() in ('l','list'):
            print(outListAll())
            
        elif userOut.lower() in ('as ','add shelf'):
            userShelf = input('номер новой полки')
            addShelf(userShelf)
            situation('directories')
        
        elif userOut.lower() in ('d','delete'):
            userDoc = input('номер документа:\n')
            print(delDoc(userDoc))
            situation()
        
        elif userOut.lower() in ('m','move'):
            userDoc = input('номер документа:\n')
            userShelf = input('номер полки:\n')
            print(moveDocInShelf(userDoc, userShelf))
            situation('directories')
        
        elif userOut.lower() == 'add':
            userDoc = input('номер документа:\n')
            userDocType = input('тип документа:\n')
            userOwnerName = input('имя владельца:\n')
            userShelf = input('номер полки:\n')
            print(addDoc(userDoc, userDocType, userOwnerName, userShelf))
            situation()
            
        elif userOut.lower() == 'situation':
            situation()
            
        elif userOut.lower() in ('q','quit'):
            break
```
